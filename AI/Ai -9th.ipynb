```python
def solve_n_queens(n):
    # To store final solutions
    solutions = []
    # The board represented as strings
    board = [["."] * n for _ in range(n)]

    # Tracking sets (The "Bound" part)
    cols = set()
    pos_diag = set() # (row + col)
    neg_diag = set() # (row - col)

    def backtrack(r):
        # Base case: successfully placed queens in all rows
        if r == n:
            copy = [" ".join(row) for row in board]
            solutions.append(copy)
            return

        for c in range(n):
            # If the square is under attack, skip it
            if c in cols or (r + c) in pos_diag or (r - c) in neg_diag:
                continue

            # Place Queen
            cols.add(c)
            pos_diag.add(r + c)
            neg_diag.add(r - c)
            board[r][c] = "Q"

            # Move to next row
            backtrack(r + 1)

            # Backtrack (Clean up)
            cols.remove(c)
            pos_diag.remove(r + c)
            neg_diag.remove(r - c)
            board[r][c] = "."

    backtrack(0)
    return solutions

# --- Run and Print ---
n = int(input("Enter N: "))
results = solve_n_queens(n)

print(f"\nFound {len(results)} solutions:")
for i, sol in enumerate(results):
    print(f"Solution {i+1}:")
    for line in sol:
        print(line)
    print()
```
