```python
import heapq

# Calculate heuristic (Manhattan Distance)
def h(board):
    dist = 0
    goal = (1, 2, 3, 4, 5, 6, 7, 8, 0)
    for i in range(9):
        if board[i] != 0:
            r, c = i // 3, i % 3
            tr, tc = goal.index(board[i]) // 3, goal.index(board[i]) % 3
            dist += abs(r - tr) + abs(c - tc)
    return dist

def solve():
    # User Input
    print("Enter 9 numbers (0 for blank):")
    start = tuple(map(int, input().split()))
    goal = (1, 2, 3, 4, 5, 6, 7, 8, 0)

    # Priority Queue stores: (f_score, g_score, current_board)
    queue = [(h(start), 0, start)]
    visited = {start}

    while queue:
        f, g, curr = heapq.heappop(queue)

        # Print current state
        print(f"\nStep {g}:")
        for i in range(0, 9, 3): print(curr[i:i+3])

        if curr == goal:
            print("\nGoal Reached!")
            return

        # Generate Neighbors
        blank = curr.index(0)
        r, c = blank // 3, blank % 3

        for dr, dc in [(-1,0), (1,0), (0,-1), (0,1)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < 3 and 0 <= nc < 3:
                neighbor = list(curr)
                target = nr * 3 + nc
                neighbor[blank], neighbor[target] = neighbor[target], neighbor[blank]
                neighbor = tuple(neighbor)

                if neighbor not in visited:
                    visited.add(neighbor)
                    heapq.heappush(queue, (g + 1 + h(neighbor), g + 1, neighbor))

solve()
```
